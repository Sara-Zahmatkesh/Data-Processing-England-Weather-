**Starting Analyses on plots of temperature, windspeed, and presseure vs. humidity and study their effects**

```python
plt.scatter (df1['Temperature (C)'], df1['Humidity'])
plt.xlabel ('Temperature (C)')
plt.ylabel('Humidity')
plt.grid ()
plt.title('Effect of the temperature on the humidity')
```

# Analysis:
* In temoeratures between -10C to 10C, humidity reaches maximum value of 100%. 
* The range of humidity change increases as the temperature increases from -20 to 15C, then its range decreases as temperaute goes up. 
* The maximum fluctuation in temperature is at 15C. 
* there are aslo some useless datapoints showing humidity is zero between -20 to -10. But they are not helpful. 


```python
plt.scatter (df1['Wind Speed (km/h)'], df1['Humidity'])
plt.xlabel ('Wind Speed (km/h)')
plt.ylabel('Humidity')
plt.grid ()
plt.title('Effect of the wind speed on the humidity')
```

# Analysis 
* As the wind speed increases, there is less data point showing the humidity values after 45 Km/h. 
* The less wind speed, the more humidity and also more humidity range fluctuation. 
* between 0 and 0.2, there is less data for humidity, showing that other factors are also playing role. 

```python
plt.scatter (df1['Pressure (millibars)'], df1['Humidity'])
plt.xlabel ('Pressure (millibars)')
plt.ylabel('Humidity')
plt.grid ()
plt.title('Effect of the air pressure on the humidity')
```

# Analysis
* It seems that the data is just over pressure 0 and almost 1000 mbars. No data is provided on the gradient of pressure. 
* this diagram is not useful for analysis, so we have to apply limit around 1oo mbar. 
* datapoints for zero pressure do not give any valuable knowledge about our data. 

```python
# datapoints are accumulated around 1000 mbar 
center_value = 1000 
# apply limit for pressure values. 
range_threshold =200  

# Filter the dataset to include only data points within the desired range
filtered_data = df1[(df1['Pressure (millibars)'] >= center_value - range_threshold) &
                    (df1['Pressure (millibars)'] <= center_value + range_threshold)]

# Plot the filtered data
plt.scatter(filtered_data['Pressure (millibars)'], filtered_data['Humidity'])
plt.xlabel ('Pressure (millibars)')
plt.ylabel('Humidity')
plt.grid ()
plt.title('Effect of the air pressure on the humidity')
```

# Analysis
* Now the plot has zoomed-in datapoints for analysis.  
* data is centeralized around 1010-1030 mbar for a range of 20% to 100% humidity. 
* in lower pressures the humidity is larger, while in more pressures we have ROUGHLY less humidity. 
* all in all by increase in pressure to 1010-1020 mbar, the range of humidity change increases. As the pressure goes up and less, we have drop in humidity and more precise values for humidity. 

# Time Analysis

```python
# using seaborn for time plot

year= pd.to_datetime (df3.Date).dt.year
import seaborn as sns
sns.set_theme(style="whitegrid")
plt.title ("humidity over years in England")
sns.lineplot(x=year, y= df3['Humidity'])
```

# Analysis
* the most humid year in England was year 2010 with humidity of 80%. & the least humid years were 2007 and 2012 with humidity value of 69%. ]
* the plot shows that the values for humidity over year is high and always more than 69% RH. 
* The diagram illustrates fluctuations of humidity, which indicates numbers of effective parameters play role on the value of humidity.
